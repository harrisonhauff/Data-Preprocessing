For both not determined and null values, we will assign the gender most like characteristics of the animal, in this case if more similar to the mean of weight and length per gender. Since we already have the mean values for both in Red squirrels and hedgehogs, we can begin their

```python
print("Male Hedgehog weight: " + str(male_H_weight))
print("Female Hedgehog weight: " + str(female_H_weight))
print("Male Hedgehog length: " + str(male_H_length))
print("Female Hedgehog length: " + str(female_H_length))
```

Comparing to both unassigned hedgehogs, they closely allign with being males, given their length, since the difference in weight was similar

```python
data.loc[(data['Animal type'] == 'Hedgehog') & (data['Gender'] == 'not determined'), 'Gender'] = 'male'
```

```python
print("Male Red Squirrel weight: " + str(male_RS_weight))
print("Female Red Squirrel weight: " + str(female_RS_weight))
print("Male Red Squirrel length: " + str(male_RS_length))
print("Female Red Squirrel length: " + str(female_RS_length))
```

Both of these characteristics are very similar between the males and females, in this case, squirrels of length 22 were decided to be female, and weight >= 0.31 kg to be male

```python
data.loc[(data['Animal type'] == 'Red Squirrel') & (data['Gender'] == 'not determined'), 'Gender'] = 'male'

data.loc[(data['Animal type'] == 'Red Squirrel') & (data['Gender'].isna()) & (data['Weight kg'] >= 0.31 ), 'Gender'] = 'male'
data.loc[(data['Animal type'] == 'Red Squirrel') & (data['Gender'].isna()), 'Gender'] = 'female'
```

Now after completing those we are left with determining the gender of the two European Bison, in this case we will only consider weight, as the difference in weight between the two undetermined is drastic

```python
male_EB_Weight = data.loc[(data['Animal type'] == 'European Bison') & (data['Gender'] == 'male'), 'Weight kg'].mean()

female_EB_Weight = data.loc[(data['Animal type'] == 'European Bison') & (data['Gender'] == 'female'), 'Weight kg'].mean()
```

```python
print("Male European Bison weight: " + str(male_EB_Weight))
print("Female European Bison weight: : " + str(female_EB_Weight))
```

There is a clear discrepency in weight like that of the not determined gender entries, these can now be adjusted

```python
data.loc[(data['Animal type'] == 'European Bison') & (data['Gender'] == 'not determined') & (data['Weight kg'] > 600), 'Gender'] = 'male'
data.loc[(data['Animal type'] == 'European Bison') & (data['Gender'] == 'not determined'), 'Gender'] = 'female'
```
